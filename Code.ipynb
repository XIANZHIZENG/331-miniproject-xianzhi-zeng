```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tkinter as tk
from tkinter import ttk
from PIL import ImageTk, Image
```

```python
# Function to create and save plot based on selected teams
def plot(team1, team2):
    # Single team plot
    if team2 == "":
        team_df = data[data["team"] == team1]  # Filter data for the selected team
        plt.figure(figsize=(5, 3), dpi=300)  # Set figure size and dpi
        sns.scatterplot(x="TRB", y="PTS", data=team_df)  # Create a scatter plot
        plt.title(f"Relationship between TRB and PTS for Players in {team1}", fontsize=8)  # Set plot title
    # Two team comparison plot
    else:
        team_df = data[data["team"].isin([team1, team2])]  # Filter data for the two selected teams
        plt.figure(figsize=(5, 3), dpi=300)  # Set figure size and dpi
        sns.boxplot(x="team", y="PTS", data=team_df)  # Create a box plot to compare teams
        plt.title(f"Distribution of PTS Between {team1} and {team2}", fontsize=8)  # Set plot title

    # Common plot settings
    plt.xlabel("Team", fontsize=6)  # Set X-axis label
    plt.ylabel("PTS", fontsize=6)  # Set Y-axis label
    plt.tick_params(labelsize=5)  # Set tick parameters
    plt.savefig("img.png", dpi=300)  # Save plot to a file
    plt.show()
```
