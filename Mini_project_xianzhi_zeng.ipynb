```python
import pandas as pd

df = pd.read_csv('lakers 2023-2024 total.csv')

df
```

```python
# Fill missing values with 0 for numeric columns
df.fillna(0, inplace=True)

# Convert 'Age' and 'GS' to integer
df['Age'] = df['Age'].astype(int)
df['GS'] = df['GS'].astype(int)

# Display the cleaned data
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualisation style
sns.set(style="whitegrid")

# 1. Points Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['PTS'], kde=True, color='blue', bins=15)
plt.title('Distribution of Total Points Scored by Players')
plt.xlabel('Total Points')
plt.ylabel('Frequency')
plt.show()

# 2. Minutes Played vs. Points
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MP', y='PTS', data=df, color='green', s=100)
plt.title('Minutes Played vs. Total Points')
plt.xlabel('Minutes Played')
plt.ylabel('Total Points')
plt.show()

# 3. Field Goal Percentage (FG%)
plt.figure(figsize=(12, 8))
sns.barplot(x='FG%', y='Player', data=df.sort_values('FG%', ascending=False), palette='viridis')
plt.title('Field Goal Percentage of Players')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Player')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('shooting lakers 2023-2024.csv')

df
```

```python
shooting_df = pd.read_csv('shooting lakers 2023-2024.csv', header=1)

# Dropping completely empty columns and rows if any
shooting_df.dropna(axis=1, how='all', inplace=True)
shooting_df.dropna(axis=0, how='all', inplace=True)

# Converting string percentages to floats and handling other numeric conversions
for col in shooting_df.columns:
    if shooting_df[col].dtype == 'object':
        try:
            # Attempt to convert percentages stored as strings to float
            if shooting_df[col].str.contains('%').any():
                shooting_df[col] = shooting_df[col].str.rstrip('%').astype('float') / 100.0
            else:
                shooting_df[col] = pd.to_numeric(shooting_df[col], errors='coerce')
        except Exception as e:
            print(f"Error converting {col}: {e}")

# Filling any NaN values that might have been introduced during conversion with 0
shooting_df.fillna(0, inplace=True)

# Display the cleaned dataframe to ensure conversions were successful
shooting_df.info()
shooting_df.head()
```

```python
# Plot 1: Average Shooting Distance for each player
plt.figure(figsize=(12, 8))
sns.barplot(x='Dist.', y='Player', data=shooting_df.sort_values('Dist.', ascending=False), palette='coolwarm')
plt.title('Average Shooting Distance for Each Player')
plt.xlabel('Average Distance (feet)')
plt.ylabel('Player')
plt.show()

# For plots 2 and 3, we need to fix the 'Player' column to show actual names instead of 0.0.
# This seems to be a mistake during the cleaning process. Let's correct this first.
shooting_df['Player'] = pd.read_csv('shooting lakers 2023-2024.csv', usecols=[1]).iloc[1:, 0].values

# Plot 2: Dunk Attempts vs. Made Dunks
plt.figure(figsize=(10, 6))
sns.scatterplot(x='%FGA', y='#', data=shooting_df, s=100, color='purple')
plt.title('Dunk Attempts vs. Made Dunks for Each Player')
plt.xlabel('Dunk Attempts (% of FGA)')
plt.ylabel('Made Dunks')
plt.grid(True)
plt.show()

# Plot 3: Corner 3-Point Shooting Efficiency
plt.figure(figsize=(12, 8))
sns.barplot(x='3P%', y='Player', data=shooting_df.sort_values('3P%', ascending=False), palette='viridis')
plt.title('Corner 3-Point Shooting Efficiency')
plt.xlabel('Corner 3-Point Shooting Percentage')
plt.ylabel('Player')
plt.show()
```
